    Sample Rate (rate):
        The sample rate determines how many data points (samples) are recorded per second during audio capture. Higher sample rates generally result in better audio quality but may require more computational resources.
        Common sample rates include 8000 Hz, 16000 Hz (default), 44100 Hz, and 48000 Hz.
        For speech recognition, a sample rate of 16000 Hz is usually sufficient. However, you can experiment with higher rates to see if they improve accuracy without significantly increasing latency or computational load.

    Chunk Size (chunk):
        The chunk size represents the number of frames per buffer that are read from the audio stream at a time. Larger chunks may reduce the number of reads from the audio device, potentially decreasing overhead, but they may also introduce more latency.
        Common chunk sizes range from 512 to 8192 frames.
        A chunk size of 1024 frames is a reasonable starting point. However, you can adjust this value based on your system's performance and the desired balance between latency and efficiency.

For optimal performance, you may need to experiment with different combinations of sample rates and chunk sizes to find the best configuration for your specific use case. Keep in mind that higher sample rates and chunk sizes may require more processing power and memory.
