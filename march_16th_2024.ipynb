```python
import pyaudio
import wave
import os
from vosk import Model, KaldiRecognizer

# Path to the directory containing the pre-trained Vosk model
model_path = r"C:\Users\poppo\Desktop\downloads firefox\vosk models\vosk-model-small-en-in-0.4"

# Initialize the Vosk model
model = Model(model_path)

# Set the sample rate
sample_rate = 16000

# Set the chunk size for reading audio data
chunk_size = 1024

# Set the duration for recording each word
duration = 5  # Adjust as needed

# Define the target words to be recognized
target_words = ["vivek", "teja", "vinod"]

# Function to record audio
def record_audio(file_name, duration):
    audio = pyaudio.PyAudio()
    stream = audio.open(format=pyaudio.paInt16, channels=1, rate=sample_rate, input=True, frames_per_buffer=chunk_size)
    frames = []
    print(f"Recording {file_name}...")
    for i in range(0, int(sample_rate / chunk_size * duration)):
        data = stream.read(chunk_size)
        frames.append(data)
    print(f"Finished recording {file_name}.")
    stream.stop_stream()
    stream.close()
    audio.terminate()
    wave_file = wave.open(file_name, 'wb')
    wave_file.setnchannels(1)
    wave_file.setsampwidth(audio.get_sample_size(pyaudio.paInt16))
    wave_file.setframerate(sample_rate)
    wave_file.writeframes(b''.join(frames))
    wave_file.close()

# Create a directory to store recorded data
data_dir = "recorded_data"
os.makedirs(data_dir, exist_ok=True)

# Record audio samples for each target word
for word in target_words:
    input(f"Press Enter to start recording '{word}'...")
    file_name = os.path.join(data_dir, f"{word}.wav")
    record_audio(file_name, duration)

# fine-tune the Vosk model using the recorded data
for word in target_words:
    file_name = os.path.join(data_dir, f"{word}.wav")
    recognizer = KaldiRecognizer(model, sample_rate)
    with open(file_name, 'rb') as audio_file:
        while True:
            data = audio_file.read(4000)
            if len(data) == 0:
                break
            if recognizer.AcceptWaveform(data):
                result = recognizer.Result()
                print(f"Recognized '{word}': {result}")

print("Fine-tuning completed.")
```

```python
# Fine-tune the Vosk model using the recorded data
fine_tuned_model_path = "fine_tuned_model"
os.makedirs(fine_tuned_model_path, exist_ok=True)

for word in target_words:
    file_name = os.path.join(data_dir, f"{word}.wav")
    recognizer = KaldiRecognizer(model, sample_rate)
    with open(file_name, 'rb') as audio_file:
        while True:
            data = audio_file.read(4000)
            if len(data) == 0:
                break
            if recognizer.AcceptWaveform(data):
                result = recognizer.Result()
                print(f"Recognized '{word}': {result}")

# Save the fine-tuned model
model.save(os.path.join(fine_tuned_model_path, "vosk-model-small-en-in-0.4-fine-tuned"))

print("Fine-tuning completed. Fine-tuned model saved.")
```
