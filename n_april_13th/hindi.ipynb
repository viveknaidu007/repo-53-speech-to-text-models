```python
import os
import pyaudio
import vosk
import json
import time

# Set the path to the Vosk model
MODEL_PATH = r"C:\Users\poppo\Desktop\github ai\vosk models\vosk-model-hi-0.22"

# Initialize the Vosk model
vosk.SetLogLevel(-1)
model = vosk.Model(MODEL_PATH)
rec = vosk.KaldiRecognizer(model, 16000)

# Function to recognize speech from audio
def recognize_audio(audio):
    start_time = time.time()  # Record start time

    rec.AcceptWaveform(audio)

    # Get the recognized text
    result = json.loads(rec.FinalResult())
    result_text = result["text"]

    end_time = time.time()  # Record end time
    latency = end_time - start_time  # Calculate latency
    
    return result_text, latency

# Function to capture audio from microphone with a duration limit
def capture_audio(duration, rate=16000, chunk=1024):
    audio = pyaudio.PyAudio()
    stream = audio.open(format=pyaudio.paInt16, channels=1, rate=rate, input=True, frames_per_buffer=chunk)
    print("Listening...")

    frames = []
    for i in range(0, int(rate / chunk * duration)):
        data = stream.read(chunk)
        frames.append(data)

    stream.stop_stream()
    stream.close()
    audio.terminate()

    return b"".join(frames)

# Main function to continuously capture and recognize live audio
def main():
    while True:
        duration = 5  # Duration in seconds
        audio_data = capture_audio(duration)

        # Recognize the captured audio
        text, latency = recognize_audio(audio_data)
        print(f"Recognized: {text} (Latency: {latency:.2f} seconds)")

if __name__ == "__main__":
    main()
```

```python
import os
import pyaudio
import vosk
import json
import time
from googletrans import Translator

# Set the path to the Vosk model and desired target language
MODEL_PATH = r"C:\Users\poppo\Desktop\github ai\vosk models\vosk-model-hi-0.22"
TARGET_LANGUAGE = "en"  # Change this to your desired language code

# Initialize the Vosk model and Google Translate translator
vosk.SetLogLevel(-1)
model = vosk.Model(MODEL_PATH)
rec = vosk.KaldiRecognizer(model, 16000)
translator = Translator()

# Function to recognize speech from audio and translate to target language (with error handling)
def recognize_and_translate(audio):
    start_time = time.time()  # Record start time

    rec.AcceptWaveform(audio)

    # Get the recognized text in Hindi
    result = json.loads(rec.FinalResult())
    hindi_text = result["text"] if result else ""

    end_time = time.time()  # Record end time
    latency = end_time - start_time  # Calculate latency for recognition

    if hindi_text:
        try:
            # Translate to target language using Google Translate
            translate_start_time = time.time()
            translated_text = translator.translate(hindi_text, dest=TARGET_LANGUAGE).text
            translate_end_time = time.time()
            translate_latency = translate_end_time - translate_start_time  # Calculate translation latency

            print(f"Recognized (Hi): {hindi_text} (Latency: {latency:.2f} seconds)")
            print(f"Translated ({TARGET_LANGUAGE}): {translated_text} (Latency: {translate_latency:.2f} seconds)")
        except Exception as e:  # Catch translation errors
            print(f"Translation Error: {e}")
    else:
        print("Could not understand audio.")

# Function to capture audio from microphone with a duration limit
def capture_audio(duration, rate=16000, chunk=1024):
    audio = pyaudio.PyAudio()
    stream = audio.open(format=pyaudio.paInt16, channels=1, rate=rate, input=True, frames_per_buffer=chunk)
    print("Listening...")

    frames = []
    for i in range(0, int(rate / chunk * duration)):
        data = stream.read(chunk)
        frames.append(data)

    stream.stop_stream()
    stream.close()
    audio.terminate()

    return b"".join(frames)

# Main function to continuously capture and recognize live audio
def main():
    while True:
        duration = 5  # Duration in seconds
        audio_data = capture_audio(duration)

        # Recognize and translate the captured audio (with error handling)
        recognize_and_translate(audio_data)

if __name__ == "__main__":
    main()
```
